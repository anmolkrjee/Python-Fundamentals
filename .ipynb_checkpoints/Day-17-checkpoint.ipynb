```python
##Hands on practice

##Create an abstract class Vehicle with an abstract method fuel_type().
##Create PetrolCar and ElectricCar subclass each implementing fuel_type() differently

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def fuel_type(self):
        pass
class PetrolCar(Vehicle):
    def fuel_type(self):
        print("Car runing on Petrol")


class ElectricCar(Vehicle):
    def fuel_type(self):
        print("Car running on electric battery")

# Create objects of subclasses

p = PetrolCar()
e = ElectricCar()

p.fuel_type()
e.fuel_type()
```

```python
##Inheritence

##Create a base class Person with name and age. Create teacher and Student subclass, each adding one extra attribute
##(e.g subject for teacher and roll number for Student). and overrifing a commmon method role(to describe who they are.


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def role(self):
        print("I am a Person")


class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def role(self):
        print("I am a Teacher. I teach", self.subject)


class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

    def role(self):
        print("I am a Student. My roll no is", self.roll_no)


t = Teacher("Anmol", 20, "DSA")
s = Student("Aman", 18, 10)

t.role()
s.role()
```

```python
##Polymorphism

##Create Square and Triangle classes, each with a draw() method that prints a different simple pattern (e.g. using *)
##Put objects of both in a lists and loop through calling draw() on each

class Square:
    def draw(self):
        print("Square Pattern : ")
        for i in range(4):
            print("*")

class Triangle:
    def draw(self):
        print("Triangle Pattern ")
        for i in range(4):
            print("*",end=" ")

list1 = [Square(), Triangle()]
for i in list1:
    i.draw()
```

```python
##Smart Library System

##Design a system to issue and return books. Store book availability safely, allow members to borrow/return books and calcualte late fees if applicable

class Library:
    
    def __init__(self, book_list = []):
        self.book_list = book_list


    
    def display_book(self):
        for i in self.book_list:
            print(i)


    def issued_book(self, book_issue):
        if book_issue in self.book_list:
            print("Book issued", book_issue)
            self.book_list.remove(book_issue)
        else:
            print("Book is not present in Library")

            
    def returned_book(self, book_return):
        print("Book returned")
        self.book_list.append(book_return)


list1 = ["DSA", "OS", "Networks", "JAVA"]
L = Library(list1)
L.display_book()
L.issued_book("DSA")
L.issued_book("Python")
L.display_book()
L.returned_book("DSA")
L.display_book()
```

```python
##Airline Ticket Reservation

##Create Economy, Business and First class tickets. Calculate ticket fare based on class and baggege weight. Print boarding detils


class Ticket:
    def __init__(self, name, destination, baggage):
        self.name = name
        self.destination = destination
        self.baggage = baggage

    def baggage_charge(self):
        if self.baggage > 15:
            return (self.baggage - 15) * 100
        return 0

    def boarding_pass(self, fare):
        print("\n----- Boarding Pass -----")
        print("Passenger :", self.name)
        print("Destination :", self.destination)
        print("Baggage :", self.baggage, "kg")
        print("Total Fare : ₹", fare)


class Economy(Ticket):
    def fare(self):
        total = 3000 + self.baggage_charge()
        self.boarding_pass(total)


class Business(Ticket):
    def fare(self):
        total = 6000 + self.baggage_charge()
        self.boarding_pass(total)


class FirstClass(Ticket):
    def fare(self):
        total = 10000 + self.baggage_charge()
        self.boarding_pass(total)


# Driver Code
e = Economy("Rahul", "Delhi", 18)
e.fare()

b = Business("Aman", "Mumbai", 12)
b.fare()

f = FirstClass("Priya", "Bangalore", 20)
f.fare()
```

```python
##Restaurant Billing

##Customers order food items. Genetrate the bill, apply GST and discounts and support Dine in and Takeway orders differently


class Restaurant:

    def __init__(self, customer, amount):
        self.customer = customer
        self.amount = amount

    def gst(self):
        return self.amount * 0.05      # 5% GST

    def discount(self):
        if self.amount > 1000:
            return self.amount * 0.10  # 10% Discount
        return 0

    def bill(self, total):
        print("\n----- Restaurant Bill -----")
        print("Customer :", self.customer)
        print("Food Amount : ₹", self.amount)
        print("Total Bill : ₹", total)


class DineIn(Restaurant):

    def total_bill(self):
        total = self.amount + self.gst() - self.discount()
        self.bill(total)


class TakeAway(Restaurant):

    def total_bill(self):
        packing_charge = 30
        total = self.amount + self.gst() + packing_charge - self.discount()
        self.bill(total)


# Driver Code

d = DineIn("Rahul", 1200)
d.total_bill()

t = TakeAway("Aman", 800)
t.total_bill()
```

```python
##Online Examination Portal

##Store student details privately. Conduct exam calculate marks, assign grades and generate a report card


class Student:

    def __init__(self, name, roll):
        self.__name = name          # Private
        self.__roll = roll          # Private
        self.marks = 0

    def conduct_exam(self, marks):
        self.marks = marks

    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 50:
            return "C"
        else:
            return "Fail"

    def report_card(self):
        print("\n----- Report Card -----")
        print("Name :", self.__name)
        print("Roll No :", self.__roll)
        print("Marks :", self.marks)
        print("Grade :", self.grade())


# Driver Code

s1 = Student("Rahul", 101)

s1.conduct_exam(88)

s1.report_card()
```
