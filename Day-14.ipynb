```python
##Build a recommendation engine that suggests products purchased by similar customers but not by the current customer


engine = {
    "Anmol" : {"Mouse", "Phone", "Laptop"},
    "Ajeet" : {"Keyboard", "Mouse", "Tablet"},
    "Aman" : {"Laptop", "Macbook", "Shoes"},
    "Ram" : {"Keyboard", "Printer", "Jeans"}
}

customer = "Anmol"

recommendations = set()

for key, value in engine.items():
    if key != customer:
        if engine[customer] & value: 
            recommendations.update(value - engine[customer])
print(recommendations)
```
