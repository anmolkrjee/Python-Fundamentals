```python
#check a number is prime or not.

num = int(input("Enter a number : "))

if num <= 1:
    print("Number is not prime")
else:
    for i in range(2,num):
        if num % i == 0:
            print("Number is not prime")
            break
    else:
        print("Number is prime.")
```

```python
#check armstrong number or not

num = int(input("Enter a number : "))

original_num = num
sum = 0;

while num>0:
    digit = num % 10;
    sum = sum + digit*digit*digit
    num = num // 10;

if (sum == original_num):
    print("Armstrong number")
else:
    print("Not Armstrong")

```

```python
#check palindrome number or not

num = int(input("Enter a number : "))
original_num = num

sum = 0
while num > 0:
    digit = num % 10
    sum = sum * 10 + digit
    num = num // 10

if sum == original_num:
    print("Palindrome number")
else:
    print("Not palindrome")
```

```python
#build an ATM 


print("Enter your ATM card in machine.")

amount = int(input("Enter how much amount in the bank."))
pin = int(input("Enter your pin : "))


if pin == 7895:
    print("Correct pin")
    entered_amount = int(input("Enter amount how much you want to withdraw."))
    if (entered_amount < 0):
        print("Please enter a valid amount.")
    elif (entered_amount <= amount):
        print("Money successfully withdraw. Please collect the cash")
        amount = amount - entered_amount
        print("Remaining amount :", amount)
    else:
         print("Entered amount is greater. Please enter a valid amount.")                    
else:
    print("Incorrect pin. Please enter correct pin.")


```
