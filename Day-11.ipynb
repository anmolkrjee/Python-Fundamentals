```python
##number of letters, digits, spaces and special characters

str = "anmol123  @@#"
letter = 0
digit = 0
space = 0
special = 0

for ch in str:
    if ch.isalpha():
        letter +=1
    elif ch.isdigit():
        digit +=1
    elif ch.isspace():
        space+=1
    else:
        special+=1
print(letter, digit, space, special)
```

```python
##find longest word in a sentence


sentence = "I am pursuing B.Tech from Lovely Professional University"

words = sentence.split()

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
print("Length:", len(longest))



sentence = "I am pursuing B.Tech from Lovely Professional University"

longest = max(sentence.split(), key = len)

print("Longest word:", longest)
```

```python
##Company payroll analytics

payroll = [100000, 50000, 60000, 70000, 75000, 65000]
print("Salary List", payroll)
print("Highest Salary", max(payroll))
print("Lowest Salary", min(payroll))
print("Total Salary", sum(payroll))
print("Average Salary",sum(payroll) / len(payroll))
```
