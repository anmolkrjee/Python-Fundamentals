```python
for i in range(1,2):
    print("Hostel", i)
    for j in range(1,3):
        print("Floor", j)
        for k in range(1,4):
            print("Room", k)
```

```python
for i in range(1,2):
    print("School", i)
    for j in range(1,3):
        print("Class", j)
        for k in range(1,4):
            print("Student", k)
```

```python
for i in range(2):
    for j in range(2):
        for k in range(2):
            if (i+j+k)%2 == 0:
                print(i,j,k)
```

```python
for i in range(3):
    for j in range(3):
        for k in range(3):
            if i == j == k:
                print(i,j,k)
```

```python
for i in range(3):
    for j in range(3):
        for k in range(3):
            if i !=j and j!=k and i !=k:
                print(i,j,k)
```

```python
##Multiplication cube using i*j*k using all possible combinations and three nested loops

for i in range(3):
    for j in range(3):
        for k in range(3):
            print(i*j*k)
```
