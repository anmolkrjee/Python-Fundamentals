```python
##3 Create a Car class . Print "Car object created " inside the constructor and "Car object destroyed" inside the destructor.
##Create two car objects and observe when the destructor messages appear


class Car:
    def __init__(self, name):
        self.name = name
        print(f"{self.name}: Car object created")

    def __del__(self):
        print(f"{self.name}: Car object destroyed")


# Create two car objects
car1 = Car("BMW")
car2 = Car("Audi")
del car1
del car2

print("Program is running...")
```

```python
##Create a rectangle class with length and width. Add two methods area and perimeter


class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        print("Area is ", self.length * self .breadth)
    def perimeter(self):
        per = 2 *(self.length+self.breadth)
        print("Perimeter is ", per)
r = Rectangle(20, 30)
r.area()
r.perimeter()
```

```python
##Create an employee class with name and salary. Add a method give_raise(percent) that updates self.salary in place (no return, just modify and
##print new salary))

class Employee:
    def __init__(self, salary, percent):
        self.salary = salary
        self.percent = percent
    def give_raise(self):
        upd = (self.salary * self.percent)/100
        new_salary = self.salary + upd
        print("New salary is ", new_salary)
e = Employee(50000, 10)
e.give_raise()
```

```python
##Encapsulation
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    @property
    def salary1(self):
        return self.__salary

    @salary1.setter
    def salary1(self, value):
        if value >= 0:
            self.__salary = value
        else:
            print("Please give correct salary")

    def add_raise(self, percent):
        upd = (self.salary1 * percent) / 100
        new_salary = self.salary1 + upd
        print("New salary is:", new_salary)


e = Employee("Anmol", 10000)

print(e.salary1)

e.salary1 = 15000
print(e.salary1)
e.salary1 = -5
e.add_raise(10)
```
