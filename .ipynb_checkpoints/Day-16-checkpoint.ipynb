```python
class Student:
    def display(self, name):
        print("Name:", name)

s = Student()
s.display("Alice")

print(type(s))

print(s.__class__) 
```

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def display_info(self):
        print(self.title, self.author, self.price)
b = Book("DSA", "Alice", 999)
b.display_info()
```

```python
class Student:
    def __init__(self, name, roll, mar):
        self.name = name
        self.roll = roll
        self.mar = mar
    def get_grade(self):
        if (self.mar> 90):
            print("A")
        elif (self.mar > 75):
            print("B")
        else:
            print("C")
s = Student("Alice", 22, 92)
s.get_grade()
```
