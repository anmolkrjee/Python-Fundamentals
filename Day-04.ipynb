```python
#count digit in a number
num = int(input("Enter a number : "))

count = 0
while num>0:
    digit = num%10
    count = count+1
    num = num//10
print(count)
```

```python
##Reverse a number
num = int(input("Enter a number : "))
reverse = 0
while num>0:
    digit = num%10
    reverse = reverse * 10 + digit
    num = num // 10
print(reverse)
```

```python
#Palindrome number or not

num = int(input("Enter a number : "))
original_num = num
reverse = 0
while num>0:
    digit = num%10
    reverse = reverse * 10 + digit
    num = num // 10

if reverse == original_num:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
#Som of digits

num = int(input("Enter a number : "))

sum = 0
while num>0:
    digit = num%10
    sum = sum+digit
    num = num//10
print(sum)
```

```python
#printing from user takes as input but it will stop when user take 0 as input

while True:
    num = int(input("Enter a num : "))
    if num == 0:
        print("Program stopped")
        break
    print(num)
```

```python
#print all numbers from 1 to 20 but skip multiples of 3

num = 1
while num <=20:
    if num%3 ==0:
        num=num+1
        continue
    print(num)
    num = num + 1
```

```python
#print numbers from 1 onwards , skip multiples of 4, stop loop when the num is greater than 20

num = 1
while num <=20:
    if num%4 ==0:
        num=num+1
        continue
    print(num)
    num = num + 1
```
