```python
##reading a file

with open("Day-15_file.txt", "r") as file:
    print(file.read())
    print("Data read")
```

```python
##data appending in a file

with open("Day-15_file.txt", "a") as file:
    file.write("Amit, 104, 9\n")
    print("Data appended")
```

```python
##File reading

with open("Day-15_file.txt", "r") as file:
    print(file.read())
```

```python
##tell and seek

with open("Day-15_file.txt", "r") as file:
    file.tell()
    print(file.read())
    file.seek(8)
    print(file.read())
```

```python
##count number of lines
count = 0
with open("Day-15_student.txt", "r") as file:
    for line in file:
        count = count + 1
print(count)
```

```python
##count number of words

count = 0
with open("Day-15_student.txt", "r") as file:
    for line in file:
        word = line.split()
        count += len(word)
print(count)
```

```python
##Search a word

search = "Anmol"
with open("Day-15_student.txt", "r") as file:
    for line in file:
        word = line.split()
        if search in word:
            found = True
            break
if found:
    print("Word exists")
else:
    print("Word does not exist")
```

```python
##copy one file to another


with open("Day-15_student.txt", "r") as source:
    content = source.read()

with open("Day-15_student_copy.txt", "w") as destination:
    destination.write(content)

print("File copied successfully.")
```

```python
##convert file content to uppercase

with open("Day-15_student.txt", "r") as file:
    for line in file:
        print(line.upper(), end="")
```

```python
##Read alternate lines

count = 0

with open("Day-15_student.txt", "r") as file:
    for line in file:
        if count % 2 == 0:
            print(line, end="")
        count += 1
```

```python
class Student:
    college = "LPU"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(self.name, self.age, self.college)

s1 = Student("Rahul", 20)
s2 = Student("Amit", 21)

s1.display()
s2.display()
```
