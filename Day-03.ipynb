```python
#Palindrome number

num = int(input("Enter a number : "))
original_num = num

reverse = 0;
while num > 0:
     last_digit = num % 10
     reverse = reverse * 10 + last_digit
     num = num // 10

if reverse == original_num:
    print("Palindrome number")
else:
    print("Not palindrome")
```

```python
#prime number

num = int(input("Enter a number : "))

if num <= 1:
    print("Not prime number")
else:
    for i in range(2, num):
        if num%i == 0:
            print("Not prime number")
            break
    else:
        print("Prime number")
```

```python
#Fibonacci series


num = int(input("Enter a number : "))
first_num = 0
second_num = 1
print(first_num)
print(second_num)
for i in range(1,num):
    next_num = first_num+second_num
    print(next_num)
    first_num = second_num
    second_num = next_num
```
