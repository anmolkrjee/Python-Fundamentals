```python
##Student_Performance_Analytics

import pandas as pd
df_student = pd.read_csv(r"C:\Users\anmol\Downloads\student_performance_project.csv")

df_student
```

```python
df_student.dropna(inplace = True)
print(df_student.isnull().sum())
```

```python
print(df_student.duplicated())
```

```python
df_student.drop_duplicates(subset = ["Name"], keep = "first", inplace = True)
```

```python
print(df_student.duplicated())
```

```python
##Filtering Students


df_student[(df_student["Math"] < 40) | (df_student["Science"] < 40) | (df_student["English"] < 40)]
```

```python
##Subject wise analysis


df_student[["Math", "Science", "English"]].agg(["max", "min", "mean"]).round(2)
```

```python
##Total, Average and ranking

df_student["Total"] = df_student[["Math", "Science", "English"]].sum(axis=1)

df_student["Average"] = df_student[["Math", "Science", "English"]].mean(axis=1).round(2)

df_student["Rank"] = df_student["Total"].rank(ascending=False)

df_student
```

```python
##pass or fail
import numpy as np 
df_student["Result"] = df_student["Average"].apply(lambda x: "Pass" if x > 40 else "Fail")

df_student
```

```python
##Gender and city wise analysis

report = df_student.groupby(["City", "Gender"])["Average"].agg(["sum", "max", "min", "mean"])
report
```

```python
##Combined group by report

report1 = df_student.groupby(["City", "Gender"]).agg({
    "Math": ["sum", "max", "min", "mean"],
    "Science": ["sum", "max", "min", "mean"],
    "Average": ["sum", "max", "min", "mean"]
})

report1
```

```python
##Topper list

top_10 = df_student[df_student["Rank"] <= 10]
top_10
```

```python
##Exporting the final report


top_10.to_csv("top_10_students.csv", index=False)
```
