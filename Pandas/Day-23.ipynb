```python
import pandas as pd
```

```python
df_student = pd.read_csv(r"C:\Users\anmol\Downloads\students_clean.csv")
df_student 
```

```python
df_student.groupby("City")["Math"].mean().round(2)
```

```python
df_student.groupby("City")["Math"].agg(["count", "min", "max", "median", "std"]).round(2)
```

```python
report = df_student.groupby("City").agg(
    total_math = ("Math", "sum"),
    avg_math = ("Math", "mean"),
    students = ("Math", "count")
).round(2)
print(report)
```

```python
df_student.groupby(["City", "Gender"])["Math"].mean().round(2)
```

```python
df_student.groupby(["City", "Gender"])["Math"].agg(["count", "min", "max", "median", "std"]).round(2)
```

```python
report1 = df_student.groupby(["City", "Gender"]).agg(
    total_math = ("Math", "sum"),
    avg_math = ("Math", "mean"),
    students = ("Math", "count")
).round(2)
print(report1)
```
