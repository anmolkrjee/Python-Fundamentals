```python
##Recursion program Factorial

def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

n = int(input("Enter a number : "))
x = fact(n)
print(x)
```

```python
##Countdown

def countdown(n):
    if n == 0:
        return
    print(n)
    countdown(n-1)
num = int(input("Enter a number : "))
countdown(num)
```

```python
##Sum of n natural number using recursion

def sum_natural(n):
    if n == 0:
        return 0
    return n + sum_natural(n-1)
num = int(input("Enter a number : "))
x = sum_natural(num)
print("Sum is ",x)
```

```python
num = int(input("Enter the number of inputs : "))
product_quantity = []
for i in range(num):
    product_quantity.append(int(input()))

square = list(map(lambda x : x**2, product_quantity))
print(square)
```

```python
##create calculator (operation) that return another function

def calc(operation):
    def add(a,b):
        return a+b
    def subtract(a,b):
        return a-b
    def multiply(a,b):
        return a*b
    def divide(a,b):
        return a/b

    if operation == "+":
        return add
    elif operation == "-":
        return subtract
    elif operation == "*":
        return multiply
    else :
        return divide

operation = input("Enter the operation : ")
calculator = calc(operation)
a = int(input("Enter a : "))
b = int(input("Enter b : "))
print(calculator(a,b))
```
