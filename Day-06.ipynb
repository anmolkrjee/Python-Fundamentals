```python
##Define a calculator 

def calc(num1, num2):
    add = num1+num2
    sub = num2 - num1
    mul = num1*num2
    div = num2/num1
    mod = num2%num1
    sq = num1*num1
    cube = num1*num1*num1
    fl_div = num2//num1
    return (add, sub, mul, div, mod, sq, cube, fl_div)

x = calc(2,11)
print(x)
```

```python
##Define a function to check the largest of two

def largest_num(num1, num2):
    if num1>num2:
        print(num1, "is largest than", num2)
    else:
        print(num2, "is largest than", num1)
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
largest_num(num1, num2)
```

```python
##Check whether the number is even or odd

def check_even_odd(num):
    if num%2==0:
        print("Even number")
    else:
        print("Odd number")
num = int(input("Enter a number : "))
check_even_odd(num)
```
