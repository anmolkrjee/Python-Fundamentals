```python
##Reverse a string

str = "Anmol"

rev = ""
for ch in str:
    rev = ch+rev
print(rev)


print(str[::-1])
```

```python
##number of vowels in a string

str = "elephant"
sum = str.count("a")+str.count("e")+str.count("i")+str.count("o")+str.count("u")
print(sum)
```

```python
##String is palindrome or not

str = "Anmol"
reverse = str[::-1]
if str == reverse:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
##total number of words in a sentence

str="I am pursuing B.Tech"
words = str.split()
print(len(words))
```

```python
##validate whether an email address has a valid format/domain

email = input("Enter your email address : ")
if email.count("@") == 1 and email.endswith("gmail.com"):
    print("Valid email")
else:
    print("Not valid email")
```

```python
##Mask a phone number except for the last four digits

phone = input("Enter phone number: ")

masked = phone[:-4].replace(phone[:-4], "*" * len(phone[:-4])) + phone[-4:]

print("Masked phone number:", masked)
```
