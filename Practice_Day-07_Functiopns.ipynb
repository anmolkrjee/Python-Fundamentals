```python
def simple_interest(time, rate, principal):
    result = (time*rate*principal)/100
    return result
time = int(input("Enter the time taken :"))
rate = int(input("Enter the rate : "))
principal = int(input("Enter the principal : "))
x = simple_interest(time, rate, principal)
print("The simple interest is ",x)
```

```python
def fahrenheit(celsius):
    result = ((celsius*9)/5)+32
    return result
celsius = int(input("Enter the temp in celsius : "))
x = fahrenheit(celsius)
print("The celsius temp is ",x)
```

```python
def bill(price, quantity, discount):
    total = price * quantity
    discount_amount = (total * discount) / 100
    final_amount = total - discount_amount
    return final_amount

price = float(input("Enter price: "))
quantity = int(input("Enter quantity: "))
discount = float(input("Enter discount (%): "))

amount = bill(price, quantity, discount)

print("Final Amount =", amount)
```

```python
def employee_salary(basic_salary, HRA_percent, DA_percent):
    hra = (basic_salary*HRA_percent)/100
    da = (basic_salary*DA_percent)/100
    total_salary = basic_salary+hra+da
    return total_salary

basic_salary = float(input("Enter Basic Salary: "))
HRA_percent = float(input("Enter HRA (%): "))
DA_percent = float(input("Enter DA (%): "))

salary = employee_salary(basic_salary, HRA_percent, DA_percent)

print("total Salary =", salary)
```
