```python
import pandas as pd
print(pd.__version__)
import numpy as np
```

```python
##Student marks analysis system

students = np.array(["Anmol", "Rahul", "Priya", "Sneha", "Amit"])

subjects = np.array(["Math", "Science", "English", "Computer"])

marks = np.array([
    [85, 90, 78, 88],
    [70, 65, 80, 75],
    [95, 92, 89, 96],
    [60, 72, 68, 70],
    [88, 84, 91, 86]
])

print(students)
print(subjects)
print(marks)
```

```python
##Average mrs per student

avg_marks = np.mean(marks, axis=1)

print("Average Marks per Student:")
for i in range(len(students)):
    print(students[i], ":", avg_marks[i])
```

```python
highest = np.max(marks, axis=1)
lowest = np.min(marks, axis=1)

print("Highest and Lowest Marks:")
for i in range(len(students)):
    print(students[i], "Highest =", highest[i], "Lowest =", lowest[i])
```

```python
##Standard deviation

std_dev = np.std(marks, axis=1)

print("Average Marks per Student:")
for i in range(len(students)):
    print(students[i], ":", std_dev[i])
```

```python
##Ranking students



rank = np.argsort(avg_marks)[::-1]

print("\nStudent Ranking:")
for i in range(len(rank)):
    print(i + 1, ".", students[rank[i]], "-", round(avg_marks[rank[i]], 2))
```

```python
##Top students above average

class_avg = np.mean(avg_marks)

print("Class Average:", round(class_avg, 2))

print("Students Above Class Average:")
for i in range(len(students)):
    if avg_marks[i] > class_avg:
        print(students[i], "-", round(avg_marks[i], 2))
```

```python
##Subject-wise Analysis

subject_avg = np.mean(marks, axis=0)

print("Subject-wise Average:")
for i in range(len(subjects)):
    print(subjects[i], ":", round(subject_avg[i], 2))
```

```python
##Percentage Calculation

total_marks = np.sum(marks, axis=1)
percentage = total_marks / 400 * 100

print("Percentage:")
for i in range(len(students)):
    print(students[i], ":", round(percentage[i], 2), "%")
```

```python
##Visualization-Ready Data

visual_data = np.column_stack((students, percentage))

print("Visualization Ready Data:")
print(visual_data)
```
