```python
import pandas as pd
```

```python
marks = pd.Series([85, 90, 78, 92])
print(marks)
```

```python
##creating a series from a dictionary

city_population = pd.Series({"Delhi" : 3380000, "Mumbai" : 20700000, "Pune" : 6800000})
print(city_population)
```

```python
##Conversion of nd array into pandas

import numpy as np

arr = np.array([10, 20, 30])
s = pd.Series(arr)
print(s)
```

```python
##Custom index

marks = pd.Series([85, 90, 78], index = ["Math", "Science", "English"], name = "marks")
print(marks)
print(marks.index)
print(marks.name)
```

```python
##Series datatypes = homogenous data

##astype() :- data type conversion

##s.values -> numpy array of values
##s.index -> the row labels
##s.dtype -> datatype
##s.shape -> tuple with number of elements
##s.name -> optional name for series
##s.size -> Total number of elements

##Methods

##s.head(n) / s.tail(n)
##s.sum(), s.mean(), s.max()
##s.sort_values()
##s.unique(), s.nunique()
##s.apply(func)
```

```python
marks = pd.Series([78, 45, 55, 96, 80, 72])
print(marks.head(3))
print(marks.tail(2))

print(marks.sort_values())
print(marks.sort_values(ascending=False))

print(marks.nunique())

print(marks.apply(lambda x : x + 10))
```

```python
##Dataframes
##Dictionary of list


data = {
    "Name" : ["Aditi", "Rohan", "Meera"],
    "Age" : [21, 23, 22],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
##List of dict

records = [
    {"Name" : "Aditi", "Age" : 21},
    {"Name" : "Rohan", "Age" : 23}
]
df = pd.DataFrame(records)
print(df)
```

```python
##Nested list dataframes 

rows = [["Aditi", 21], ["Rohan", 23]]
df = pd.DataFrame(rows, columns = ["Name", "Age"])
print(df)
```

```python
##Creating dataframes from numpy array

arr = np.array([[1, 2], [ 3,4]])
df = pd.DataFrame(arr, columns = [ "A", "B"])
print(df)
```
