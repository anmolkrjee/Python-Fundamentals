```python
##Aceept 5 strings from user and store them as part of tuple

tup = ()
for i in range(5):
    strings = input("Enter a string : ")
    tup = tup + (strings, )
print(tup)

for i in range(5):
    print("tup[",i,"] = ",tup[i])
```

```python
##Dict

dict1 = {}
student = {
    "Name":"Anmol",
    "Age":20,
    "CGPA":7.9
}
print(student["Name"])
##print(student["Phone"])  ##Keyerror

##Get method

print("Using get method:-", student.get("Name"))
print(student.get("Phone")) ##None
print(student["Age"])

student["city"] = "Delhi" ##ADD KEY AND VALUE
student["Age"] = 23 ##UPDATE


student.update({"Semester":5})  ##using .update method

print(student.pop("Semester"))  ##Key is passed that will be deleted
print(student.popitem())  ##delete the last key and value
print(student) 
```
