```python
balance = 1000
def deposit():
    global balance
    balance = balance + 500
    print(balance)

def withdraw():
    global balance
    balance = balance - 300
    print(balance)
deposit()
withdraw()

print(balance)
```

```python
##Sum of two numbers using lambda function

sum = lambda x, y: x+y
print(sum(7, 8))
```

```python
##Multiplication of two numbers using lambda function

multiply = lambda x, y: x*y
print(multiply(7, 8))
```

```python
##Largest number

from functools import reduce
numbers = [87, 56, 45, 81, 12]
largest_num = reduce(lambda x, y : x if x > y else y, numbers)
print(largest_num)
```

```python
##Even or odd
num = int(input("Enter a number : "))
even_odd = lambda x : "even" if x %2==0 else "odd"
print(even_odd(num))
```

```python
##Positive and negative

num = int(input("Enter a number : "))
positive_negative = lambda x : "Positive" if x > 0 else "Negative"
print(positive_negative(num))
```

```python
## Apply discount of 20% on every product

price = [1000, 1200, 1400, 1600, 1800, 2000]

dis = list(map(lambda x : x - ((x*20)/100), price))
print(dis)
```

```python
## taking a number and incrementing 10 times like n, n+1, n+2..........till n+10

num = int(input("Enter a number : "))

increment_number = map(lambda x : x+1, num)
print(list(increment_number(num)))
```
